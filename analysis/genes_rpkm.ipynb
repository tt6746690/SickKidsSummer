```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
GENES_RPKM = "../data/GenesRPKM.txt"
GENE_SYMBOL =["DDX11L1","WASH7P"]
TISSUES =["Bladder"]
```

```python
genes_rpkm = pd.read_table(GENES_RPKM,skiprows=2)
del genes_rpkm["Name"]

```

```python
genes =list (genes_rpkm["Description"])
tissue =list ( genes_rpkm.columns)

data=[]
tissue.pop(0)
rows =[]
for x in range (len (tissue)):
   
    rows.append ( list (genes_rpkm.loc[x]))
for x in rows:
    x.pop(0)

```

```python

indx,plot_data,indx2,final_data,final_data2 =[],[],[],[],[]
for x in GENE_SYMBOL:
    indx.append (genes.index (x))

for x in indx:
    
    plot_data.append( rows[x])
for y in TISSUES:
    indx2.append (tissue.index(y))
    
for data in plot_data:
    final_data=[]
    for y in indx2:
        final_data.append (data[y])
    final_data2.append (final_data)
    

  
```
